## Grading
- How to ensure fairness:
    - Instead of grading all submissions end-to-end, grading the same subsets of the submission for all of the submissions at the same time
    - When a grading task is completed, revisit the earliest submissions to ensure they received the same scrutiny as later submissions
    - Keep a grading log as you work (next slide)
    - If multiple people are grading the submissions, either
        1. Have the same person do the same tasks for all the submissions
        2. Meet after grading some number of submissions to discuss initial impressions and ensure everyone grades the submissions they handle in comparable ways

## Assisting Students One-on-One
- Many teachers, early in their career, are excited to have all the answers. This is natural.
- But this is not actually what students need. They need someone who can listen to them talk about the problem, diagnose their misunderstandings, then help them make progress towards building or repairing their mental architecture.
- If you don't help them fix the gaps and breaks in their understanding of the topic, they're doomed to keep making the mistake.
- So let's break this down to a protocol:
    1. Ask the student to explain, in their own words, what the difficulty is (delaying introducing the computer as long as possible). It is essential that students learn to communicate their problems without shoving a computer in someone's face.
    2. Engage with what the student is saying and try to diagnose the issue in your own head: What is the student misunderstanding? What context have they missed?
    3. **Refer them to the relevant course material.** (e.g. "I see you're having a problem with determining when a matrix is invertible, let's find that in the notes/book/slides.")
    4. Talk through their problem with them in light of the course material, using your diagnosis to guide the discussion. Help them try to resolve the issues themselves, given the resources available.
    5. Get out the computer or a pad of paper and start working on an explicit solution.
- As you go along, it helps to say things like, "It sounds like you're having trouble with X. Does that sound right?" It makes the student feel like they are being heard, and you get more information.
- When it comes to code, students will want to "just fix it". You will probably feel the same way: "Oh, they have the wrong file path." But the thing is, unless it's really just a typo, they will continue to miss something fundamental and continue to make these mistakes, forever. 

## Diagnosing Student Misunderstanding
- Putting aside genuine mistakes, how do you diagnose student misunderstanding?
- It's helpful to think about the space of questions, $Q$, and space of answers, $A$
- The student is trying to answer a question $q$ and has some sort of answer $a'$
- Unfortunately, $a'$ is not the answer to $q$
- Often, it helps to ask, "What is $q'$, the question the student actually attempted to answer, and how do $q$ and $q'$ differ? **How/why** did the student end up working on $q'$ instead of $q$?"
- In numerical analysis, we call this **backward error** (distance between true $q$ and implied $q'$ questions) rather than **forwards error** (distance between true $a$ and provided $a'$ answers)
- So, **Why did the student answer the wrong question?** instead of **Why did the student get the wrong answer?**

## Diagnosing Student Misunderstanding

So, what was the student trying to do, and what did they actually do:

1. A student is working on a prediction problem where the majority class has a value zero and the minority class has the value 1. The assignment asks them to do a train-test split, use the training set to fit a model to predict the probability that a case results in a 1, make predictions on the test set, and report the $R^2$ for the test set. The accuracy of their model is around .9, with which they are rather pleased. What went wrong? What are some helpful $q'$?
2. A student is attempting to take the derivative of absolute value at zero. The student asserts the correct answer is 0, because 0 is indeed a tangent line to the graph at zero. What went wrong? What are some helpful $q'$?
3. A student is attempting multivariate linear regression. For each variable, they compute $b_k = \text{cov}(x_k,y)/\sigma_k$ and set the intercept equal to $\bar{y}$. What went wrong? What are some helpful $q'$?
4. A student is using a data driven approach to pick a hyperparameter; say, the max depth of the trees for a random forest. The assignment says to consider max depths of 5, 6, and 7. That's 3 possibilities, so the student asserts that they should do 3-fold cross validation. What went wrong? What are some helpful $q'$?
5. A student is asked "Consider the statement 'War is never Pareto Optimal.' Is this True, False, or Uncertain, and explain." The student claims that it's necessary that some wars are fought, like the American civil war or the Ukrainian war, so the statement must be false. What went wrong? What are some helpful $q'$?

- This is where being an expert yourself matters. Being able to listen to someone else and figure out where they went wrong is more than just being able to get the right answer.
- So what you'll find is that thinking about $q'$ means you're thinking about what tragic missteps might have brought this poor student to this point, where focusing on $a$ as the answer to $q$ doesn't really spark any ideas about what's going on in the student's head.
- This is also a way of supporting students while still acknowledging that what they did is wrong. I am attracted to data science because there's a sense of curiosity and play that is dead in many fields. Don't kill students' curiosity. Thinking about and discussing $q'$ can be a great way to introduce them to other ideas and reward them for thinking, even if they didn't get the right answer. 

## Helping Students with Math One-on-One
- Many students are somewhat traumatized by bad math experiences
- There are three techniques that work really well with helping students understand math:
    - Instructive counter-examples: If $A, B \rightarrow C$, find examples of not $A$ and not $B$ leading to not $C$ to illustrate why they are necessary for $C$
    - Parameterized example: Use a simplified version of the original idea, with some variables left in to illustrate a specific point 
    - Simulation: Your advantage as a data scientist is that you can use the computer to play and illustrate.

## Helping Students with Math: Instructive Counter-example
- Instructive counter-examples work best when you have a complex idea that relies on a set of complicated assumptions.
- Everyone here has had calculus and you'll have to optimize lots of functions as a data scientist. Here's a great result, the Weierstrass Theorem: "Every continuous function mapping $S \rightarrow \mathbb{R}$ where $S$ is a closed and bounded (compact) subset of $\mathbb{R}^N$ has a maximum and a minimum." 
- The proof is clever and requires using subsequences of bounded sequences to construct a sequence $(x_n,y_n)$ that converges to the supremum or infimum of the function on $S$. It's not the kind of thing you'd expect anyone to magically intuit about the natural of reality.
- So instead, let's unpack these conditions:
    - Why continuous functions? (Why might discontinuities be a problem?)
    - Why closed domains? (What might happen with a continuous function on the boundary of, say, $(0,1)$?)
    - Why bounded domains? (What might happen if you're optimizing over $[0,\infty)$?)
- It turns out that once you construct instructive counter-examples, you actually have a much better understanding of the theorem itself (and the proof makes more sense: it encourages a topological view of the problem, where you're mapping a closed/bounded $S$ by a continuous $f$ into a closed, bounded set, which must have a maximum and minimum.).

## Helping Students with Math: Parameterized Example
- Parameterized examples work best when the topic is too broad for a student to understand completely in the abstract.
- For example, a lot of students use the result: "If $x^*$ is a maximizer or minimizer of a differentiable function $f(x)$ mapping $(c,d)\rightarrow \mathbb{R}$, then $f'(x^*)=0$"
- Students often internalize this as, "To find a maximizer: Take a derivative, set it equal to zero, and solve!" Well, sort of. Let's help disabuse a student of this confusion.
- In order to illustrate some of the issues, just start with a quadratic: $f(x) = g - ax + bx^2$
- The first-order necessary condition is
$$
- a + 2bx^* = 0.
$$
- Let's analyze this carefully:
    - If $b=0$, there are no critical points because we can't solve the equation for $x^*$: It's a linear function on an open set $(c,d)$, and the maximizer/minimizer fails to exist (see previous slide). These ideas really only work if the function has some curvature to it, or else you're working on a Linear Programming Problem.
    - If $b\neq 0$, we can solve for $x^* = a/2b$. If that is in the interval $(c,d)$, there's a maximizer or minimizer, but we're not sure which one; otherwise, $f(x)$ is strictly increasing or decreasing on $(c,d)$ and there's no maximizer or minimizer.
    - If we plug $x^* = a/2b$ back into the objective function, we get
    $$
    f(x^*) = g + x^*(bx^* -a  ) = g - \dfrac{a^2}{4b}
    $$
    So the sign of $a$ ends up being irrelevant. If $b>0$, we must be at a minimum, since the sign of second term becomes negative. Conversely, if $b<0$, the sign of the second term becomes positive, and we must be at a maximum. 
- This example works because the parameters secretly illustrate important ideas without invoking them explicitly (convexity, concavity, second-order sufficiency), and the parameters allow you to illustrate the ideas in one clear example with a single real variable. Everything is easy to plot, students are already familiar with quadratic forms, and the abstract nature of the optimization framework becomes more approachable.
- These kinds of little math poems are really valuable for students to build intuition. When students are failing to grasp the abstract concept, make it concrete, with the specific misunderstanding mapped into a parameter of the model.

## Helping Students with Math: Simulation
- Simulation is a great way to help students understand math topics, particularly statistical ones.
- The Central Limit Theorem is probably one of the more misunderstood concepts at the undergraduate level. Many students mentally confuse it with the definition of a Normally distributed random variable or the weak law of large numbers or a version of the delta method.
- To explain the CLT, it helps to have a slightly weird example that moves away from a normally distributed experiment.
- Imagine you're doing quality control for a lightbulb company. You are trying to determine the average lifespan of a bulb, to see if it passes Federal guidelines. You test 200 lightbulbs, simply running them until they burst. What does their average lifetime look like?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(100)
 
theta = 1/1000 # A bulb's expected lifetime is 1k hours
N = 200
u = np.random.rand(N)
t = - np.log(1-u)/theta # Sample of bulbs

print( 'Mean lifetime: ', np.mean(t) )
sns.histplot(t)
```

- So some bulbs are essentially defective and burst well below 1000 hours, but many hang on for very long runs of up to 7000 hours.
- Now, this number: 1049.48. What do we think of this number? It's random, a function of the sample. Is it precisely estimated?
- Hypothetically, let's run this experiment many, many times and compute this number for each experiment. What do we get? Let's represent the results using a kernel density plot:

```python
for sample_size in [3,10,100]:
    means = [ np.mean( - np.log(1-np.random.rand(sample_size))/theta ) for i in range(1000)]
    sns.kdeplot(means, label = 'N: ' + str(sample_size))
plt.legend()
plt.show()
```

- So if we think of each lightbulb as an independent experiment, and let the sample size get very large, the density of simulated means seems to be concentrating on something. This is essentially a numerical illustration of the WLLN: The variance of $\bar{X}_N = \sigma^2/N$ is converging to zero around the true mean.
- Notice that by showing this first, the centering and $\sqrt{N}$ normalization will now make more sense to the student: You can even change these features and see what happens
- Where can we see the CLT? A standard statement of the CLT is: "Let $X_1, X_2, ..., X_N$ be a sequence of iid random variables sharing a distribution with mean $\theta$ and variance $\sigma^2 < \infty$. Then $\sqrt{N} ( \bar{X}_n - \theta)$ converges in distribution to a Normal with mean 0 and variance $\sigma^2$."
- Let's illustrate that claim clearly:

```python
for sample_size in [3,10,100]:
    means = [ np.sqrt(sample_size) * (np.mean( - np.log(1-np.random.rand(sample_size))/theta ) - 1/theta) for i in range(1000)] #1/theta since exponentially distributed RV
    sns.ecdfplot(means, label = 'N: ' + str(sample_size))
plt.legend()
plt.show()
```

- This is an obvious example, but shows how powerful simulation can be in illustrating issues and resolving student misunderstandings
- In the above plot, we used the ECDF instead of the PDF. Why make that choice?
- A piece of why it works is that the lightbulb example doesn't have any obvious Normality built into it; some students get the idea that the CLT is about mapping of Normals preserving Normality
- Simulation isn't a replacement for math, but it can be a powerful complement.

## Helping Students with Code One-on-One
- Students often send me a single line of code and ask, "What is the problem?" These kids.
- It's very hard to help students without being able to see their code and run it yourself.
- To work with students on code, I do the following:
    1. Make sure their work is on GitHub or somewhere you can actually access it
    2. Clone their repo (I usually do this on Google Colab), and you take over the typing
    3. Sit with them and have them talk through the problem code, line by line, so they can explain why they made the choices they did
    4. When you find the error or misunderstanding, comment out the incorrect lines and work collaboratively to fix the problem
    5. **Restart the kernel and run the code again from the top: Does it really work?**
    6. Send them the commented out lines and the solution in an email
- I think we should consider it very impolite to touch other people's computers: Make a habit of opening their code and projects on your computer, and controlling the conversation.
- Students need you to slow them down and make them talk through their thought processes. If you are the hands and they are the brain, they are more effective at thinking. This is facilitated by doing the work on your computer.
- Typically, the student will discover the error as they talk through the problem with you, for which you should praise them, even if it's just a silly typo. 
- When you've addressed their particular problem, you can do humanity a service by offering other, general suggestions:
    1. Repackage code as a function if you're using it repeatedly
    2. If you're using Jupyter or RStudio, use small code chunks to isolate errors
    3. Vectorize code for performance and readability
    4. Add comments to your code so your future self can understand what you were thinking

## Scaling Up: Review Sessions or Sections
- "How do you help students at scale?" is the question of running a class or section. 
- The same themes apply from one-on-one, but the issue is that the students are now 
    1. Providing positive externalities for one another by taking risks and asking questions, which is great
    2. Socially performing for one another, which can be unfortunate
- UVa students are outrageously risk averse. They do not want to raise their hands, they do not want to ask questions, they especially do not want other students to see their code ("code-shaming")
- I suspect this is -- maybe subconsciously -- because of social media: There is a small but non-zero chance that each moment of your life is being recorded, and if it becomes the worst moment of your life, it can go viral and define you as a person.
- A strategy that sometimes works is vulnerability: "Hey everyone, I'm going to live code this up like a donkey! See, everyone struggles!" This can harm your credibility, especially if you're a female instructor (which I loathe, of course). If students view your mistakes as humanizing, good; if they view them as evidence of your incompetence, bad (e.g. "The TA can't even solve the problems!@!##!").
- I often pick a problem I think has hidden depth, and kind of copy-paste my solutions into a notebook and talk students through the output. This can get them engaged, and open up a bit. Starting a session that way can help people loosen up.
- You want the first few sessions to be lively or interesting. If they're not, students probably won't come back. I'd come prepared to just do a few problems. Sometimes, doing a quick review of the recent classes is enough value added that students will come back for that.
